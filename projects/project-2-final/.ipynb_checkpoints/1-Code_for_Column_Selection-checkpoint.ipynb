```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from scipy import stats

from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.datasets import load_breast_cancer, load_digits, load_iris
from sklearn.metrics import mean_squared_error

from sklearn.compose import make_column_transformer
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import PolynomialFeatures, OrdinalEncoder

from category_encoders import OneHotEncoder

from sklearn.pipeline import make_pipeline
import sklearn
sklearn.set_config(display = 'diagram')
from sklearn.feature_selection import RFE

pd.options.display.max_rows = None
```

```python
df = pd.read_csv('./datasets/train.csv')
df.head()
```

```python
df.info()
```

```python
# Dropping columns with too many NaNs

new_df = df.drop(['Alley', 'Lot Frontage', 'Pool QC', 'Misc Feature', 'Fireplace Qu', 'Fence', 'Garage Type', 'Garage Yr Blt', 'Garage Finish'], axis = 1)
```

```python
dropped_df = df[['Alley', 'Lot Frontage', 'Pool QC', 'Misc Feature', 'Fireplace Qu', 'Fence', 'Garage Type', 'Garage Yr Blt', 'Garage Finish']]
dropped_df.info()
```

```python
# From data dictionary, these columns are ordinal. Using mapping function to match score in range 0-5

new_df['Bsmt Qual'] = new_df['Bsmt Qual'].map({np.nan: 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5})
new_df['Garage Qual'] = new_df['Garage Qual'].map({np.nan: 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5})
new_df['Garage Cond'] = new_df['Garage Cond'].map({np.nan: 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5})
new_df['Lot Shape'] = new_df['Lot Shape'].map({'IR3': 0, 'IR2': 1, 'IR1': 2, 'Reg': 3})
```

```python
# Drop remaining NaN rows

new_df = new_df.dropna()
```

```python
# One Hot Encode in order to display p-values

ohe = OneHotEncoder()
```

```python
X = new_df.drop(['SalePrice'], axis = 1)
y = new_df['SalePrice']
```

```python
ohe = OneHotEncoder(use_cat_names = True)
```

```python
X_encoded = ohe.fit_transform(X)
```

```python
# fit model in order to find p-values

model = sm.OLS(y, X_encoded).fit()
```

```python
# Find p-values for all columns

pd.set_option('display.float_format', lambda x: '%.3f' % x)
model.pvalues.sort_values(ascending = True)
```

```python
new_df.corr()['SalePrice'].sort_values(ascending=False)
```

Based on p-values (0.01 or lower) and correlation to SalePrice (above 0.6), the following would be the best selections:<br>

1) Gr Liv Area: p-value = 0.000, corr = 0.701<br>
2) Overall Qual: p-value = 0.000, corr = 0.798<br>
3) Total Bsmt SF: p-value = 0.000, corr = 0.629<br>
4) 1st Flr SF:  p-value = 0.009, corr = 0.630<br>
5) Garage Area: p-value = 0.010, corr = 0.651<br>

```python
# Check for multicollinearity between top features

plt.figure(figsize = (10, 8))
sns.heatmap(new_df[['Gr Liv Area', 'Overall Qual', 'Total Bsmt SF', '1st Flr SF', 'Garage Area']].corr(), annot= True);
```

Conclusion: 1st Flr SF is too similar to Total Bsmt SF --> only keeping Total Bsmt SF because it's p-value is lower.
