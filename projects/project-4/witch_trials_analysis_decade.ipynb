```python
battles_clean = battles[battles['gadm.adm2'].notna()]
battles_clean.info()
```

```python
df = pd.merge(left = trials_clean, right = battles_merge, on = 'decade_loc', how = 'left')
df.head()
```

```python
df.info()
```

```python
df['battle'] = np.where(df['battle'].isna(), 0, 1)
df['battle'].value_counts()
```

```python
df.head()
```

There are a lot of trials that have no battle associated. I think it will help to add a lag of 1-5 years because some of the effects may not be immediate.

```python
battles_lag = battles_clean[['battle', 'decade', 'gadm.adm2']]
```

```python
lag_battles = {i: f'lag_{i}' for i in range(1, 6)}
lag_battles
```

```python
for i, column_name in lag_battles.items():
    battles_lag[column_name] = (battles_lag['decade'] + i*10).astype(str) + battles_lag['gadm.adm2']
    
    
battles_lag
```

```python
battles_lag = battles_lag.drop(columns = ['battle', 'decade', 'gadm.adm2'])
```

```python
columns = battles_lag.columns
columns
```

```python
df_trials = df.copy()
```

```python
for column in columns:
    lag = battles_lag[[column]]
    temp_df = pd.merge(left = df_trials, right = lag, left_on = 'decade_loc', right_on = column, how = 'left')
    temp_df[column] = np.where(temp_df[column].isna(), 0, 1)
    
    df_trials[column] = temp_df[column]
```

```python
df_trials
```

```python
df_trials['lag_1'].value_counts()
```

```python
df_trials['lag_5'].value_counts()
```

```python
df_trials.info()
```

```python
df_trials.drop(columns = ['decade_loc'], inplace=True)
```

```python
X = df_trials.drop(columns = 'tried')
y = df_trials['tried']
```

```python
X = pd.get_dummies(data = X, columns = ['gadm.adm2', 'gadm.adm1', 'gadm.adm0', 'record.source'], drop_first = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 8)
```

```python
X.shape
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```
