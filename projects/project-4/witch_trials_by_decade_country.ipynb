```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, BatchNormalization
from tensorflow.keras.layers import Dropout
```

### Read in Trials Data

```python
trials = pd.read_csv('./witch-trials-master/data/trials.csv')
trials.head()
```

```python
trials.info()
```

```python
trials_clean = trials.copy()
```

```python
trials_clean.isna().sum()
```

#### Create column to connect trials + battles

Tried using year + most descriptive location, but received bad results.

Here I am using the decade and the country.

```python
trials_clean['decade_loc'] = trials_clean['decade'].astype(str) + trials_clean['gadm.adm0']
trials_clean.head()
```

### Read in Battles Data

```python
battles = pd.read_csv('./witch-trials-master/data/battles.csv')
battles.head()
```

```python
battles.info()
```

```python
battles_clean = battles.copy()
```

#### Make same connecting column

```python
battles_clean['decade_loc'] = battles_clean['decade'].astype(str) + battles_clean['gadm.adm0']
battles_clean.head()
```

#### Create count of battles per decade/location

```python
battle_count = pd.DataFrame(battles_clean['decade_loc'].value_counts())
battle_count = battle_count.reset_index()
battle_count.rename(columns = {'index': 'decade_loc', 'decade_loc': 'count_of_battles'}, inplace = True)
battle_count
```

```python
#drop uneeded columns
trials_clean = trials_clean.drop(columns = ['city', 'lon', 'lat', 'deaths'])
```

```python
trials_clean.info()
```

#### Merge battles with trial data

```python
df = pd.merge(left = trials_clean, right = battle_count, on = 'decade_loc', how = 'left')
df.head()
```

```python
#since it is a count of how many battles, I will fill all N/As with zeros
df['count_of_battles'] = df['count_of_battles'].fillna(0)
```

```python
df.head()
```

```python
#drop more uneeded columns
df.drop(columns = ['year', 'gadm.adm2'], inplace=True)
```

```python
#drop N/As
df.dropna(inplace = True)
```

```python
plt.boxplot(df['tried']);
```

```python
sns.pairplot(df, corner = True)
```

```python
df[df['tried'] > 350]
```

```python
df.groupby(by = 'record.source').sum()
```

### Build first model

```python
X = df.drop(columns = ['tried', 'decade_loc'])
y = df['tried']
```

```python
X = pd.get_dummies(data = X, columns = ['gadm.adm1', 'gadm.adm0', 'record.source'], drop_first = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 15)
```

```python
#baseline RMSE
baseline_preds = np.full_like(y_test, np.mean(y_test))
mean_squared_error(y_test, baseline_preds, squared = False)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
lr.score(X_train, y_train)
```

```python
preds = lr.predict(X_test)
```

```python
mean_squared_error(y_test, preds, squared = False)
```

### Feature engineering

There are a lot of trials that have no battle associated. I think it will help to add a lag of 1 decade years because some of the effects may not be immediate.

```python
lag = battles_clean.copy()
```

```python
lag['lag_10'] = lag['decade'] + 10
lag.head()
```

```python
lag['decade_loc_lag'] = lag['lag_10'].astype(str) + lag['gadm.adm0']
lag.head()
```

```python
battle_lag = pd.DataFrame(lag['decade_loc_lag'].value_counts())
battle_lag = battle_lag.reset_index()
battle_lag.rename(columns = {'index': 'decade_loc', 'decade_loc_lag': 'count_of_battles_lag'}, inplace = True)
battle_lag
```

```python
df_trials = df.copy()
```

```python
df_trials = pd.merge(left = df_trials, right = battle_lag, on = 'decade_loc', how = 'left')
df_trials
```

```python
df_trials['count_of_battles_lag'] = df_trials['count_of_battles_lag'].fillna(0)
```

```python
df_trials
```

```python
X2 = df_trials.drop(columns = ['tried', 'decade_loc'])
y2 = df_trials['tried']
```

```python
X2 = pd.get_dummies(data = X2, columns = ['gadm.adm1', 'gadm.adm0', 'record.source'], drop_first = True)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, random_state = 15)
```

```python
lr2 = LinearRegression()
lr2.fit(X2_train, y2_train)
lr2.score(X2_train, y2_train)
```

```python
lr2.score(X2_test, y2_test)
```

```python
preds2 = lr2.predict(X2_test)
```

```python
mean_squared_error(y2_test, preds2, squared = False)
```

```python
predictions = pd.DataFrame(y2_test)
predictions['preds'] = preds2

predictions.rename(columns={'tried': 'Actual Trials', 'preds': 'Predicted Trials'}, inplace = True)
predictions.head()
```

```python
plt.figure(figsize = (7, 5))
sns.regplot(data = predictions, x = 'Actual Trials', y = 'Predicted Trials', line_kws={'color': 'black'})
plt.title('Actual vs. Predicted Trials')

;
```

### Decision Tree & Random Forest

```python
tree = DecisionTreeRegressor()
tree.fit(X2_train, y2_train)
tree.score(X2_train, y2_train)
```

```python
tree.score(X2_test, y2_test)
```

```python
pipe = make_pipeline(RandomForestRegressor())
```

```python
pipe.get_params().keys()
```

```python
params = {'randomforestregressor__n_estimators': [50, 100, 200],
         'randomforestregressor__max_depth': [None, 50, 25]}
```

```python
gs = GridSearchCV(pipe, params, n_jobs = -1)
```

```python
gs.fit(X2_train, y2_train)
```

```python
pd.DataFrame(gs.cv_results_).sort_values(by = 'rank_test_score')
```

```python
pipe = make_pipeline(StandardScaler(), RandomForestRegressor())
pipe.fit(X2_train, y2_train)
pipe.score(X2_train, y2_train)
```

```python
pipe.score(X2_test, y2_test)
```

```python
pipe_preds = pipe.predict(X2_test)
mean_squared_error(y2_test, pipe_preds, squared = False)
```

### Conclusion

The best model was Random Forest, bring the RMSE down to 9.36. When looking at feature importances, the count of battles and lagged count of battles end up in the top 10 out of 162 features. Some time periods, historians, and places are showing up as important as well. Decade and century hold the highest importance.
